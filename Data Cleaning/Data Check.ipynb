```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('covid_19_data.csv')
df.describe().transpose()
```

```python
df
```

```python
# Consider a column, is it numeric or categorical?
# Numeric then check the range and distribution
# Categorical then check number of classes and distribution
```

```python
# Numeric range
print('Range from', min(df['Confirmed']), 'to', max(df['Confirmed']))
```

```python
# Numeric data
sns.set_style("whitegrid")
sns.distplot(df['Confirmed'], bins=range(int(min(df['Confirmed'])), int(max(df['Confirmed'])), 1000), kde = False)
plt.show()
```

```python
# Notice value between 0 to 1000 overpower the column
```

```python
sns.scatterplot(x = df.index, y = df['Confirmed'], hue = df['Country/Region'])
plt.show()
```

```python
# Notice too df['Country/Region'] has too many classes
# Reduce them so it is readable
df_reduced = df[(df['Country/Region'] == 'Mainland China') | 
                (df['Country/Region'] == 'US') | 
                (df['Country/Region'] == 'Italy') | 
                (df['Country/Region'] == 'Iran')]
# Already sorted by date, so using index for x-axis
sns.scatterplot(x = df_reduced.index, y = df_reduced['Confirmed'], hue = df_reduced['Country/Region'])
```

```python
# Notice US and China has multiple lines spanning out
# Hypothesis: Infection rate vary depending on sub-regions for the US and China. Perhaps due to larger area of country.
```

```python
# Already sorted by date, so using index for x-axis
sns.scatterplot(x = df_reduced.index, y = df_reduced['Deaths'], hue = df_reduced['Country/Region'])
```

```python
# Notice China plateaus, as the other 3 continue to grow in deaths
```

```python
# Already sorted by date, so using index for x-axis
sns.scatterplot(x = df_reduced.index, y = df_reduced['Recovered'], hue = df_reduced['Country/Region'])
```

```python
# Notice China plateaus, as the other 3 continue to grow in recoveries
# Confirmed is a superset of deaths and recoveries, notable is that deaths and recoveries follow the same pattern
```
